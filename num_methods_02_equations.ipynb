**Ex 01. Equation**

```python
import numpy as np
from scipy.optimize import fsolve

def eq(x):
  eq = 2 * x - 4
  return eq

init_guess = [1]
roots = fsolve(eq,init_guess)
print(roots)
np.isclose(eq(roots),[0])
roots = np.array(roots, dtype='int16')
print(roots)
```

```python
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 1000)
y = eq(x)

plt.figure(figsize=(6, 4))
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

**Ex 02. Equation**

```python
import numpy as np
from scipy.optimize import fsolve

def eq(x):
  eq = x * np.cos(x) - 4
  return eq

init_guess = [-100]
roots = fsolve(eq,init_guess)
print(roots)
np.isclose(eq(roots),[0])
```

```python
import matplotlib.pyplot as plt
x = np.linspace(-150, 100, 1000)
y = eq(x)

plt.figure(figsize=(6, 4))
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

**Ex 03. Systems of equations**

```python
from scipy.optimize import fsolve

def eq(vars):
  x, y = vars
  eq_1 = x**2 + y**2 - 4
  eq_2 = x**2 - y - 1
  return [eq_1, eq_2]

init_guess = [1,1]
roots = fsolve(eq, init_guess)
print("Solution to the system:", roots)

x_0, y_0 = roots
print(f"Solved values are x = {x_0:.2f} and y = {y_0:.2f}")

print("Verification:")
print(f"eq_1(x, y) = {x_0**2 + y_0**2 - 4:.2f}")
print(f"eq_2(x, y) = {x_0**2 - y_0 - 1:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 400)
y = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x, y)

# Define the equations for plotting
Z1 = X**2 + Y**2 - 4
Z2 = X**2 - Y - 1

plt.figure(figsize=(6, 4))
plt.contour(X, Y, Z1, levels=[0], colors='blue')
plt.contour(X, Y, Z2, levels=[0], colors='red')
plt.plot(x_0, y_0, 'go', label='Solution')
plt.xlabel('x')
plt.ylabel('y')
plt.title('2D Visualization of Nonlinear Equations')
plt.legend()
plt.grid(True)
plt.show()
```
